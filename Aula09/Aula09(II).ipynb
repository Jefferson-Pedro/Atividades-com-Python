#Coleta de dados de uma pagina HTML
##Objetivo: Coletar os dados da tabela de criptomoedas.

###Verificando o Status do site

```python
import pandas as pd
import requests

url = 'https://www.worldcoinindex.com/'
crypto_url = requests.get(url) #verifica o status do site
print('Status da pagina: ', crypto_url)
```

###Mostrando as informações da tabela no site

```python
import pandas as pd
import requests

crypto_data = pd.read_html(crypto_url.text)

crypto_data
```

###Processo de limpeza dos dados

```python
len(crypto_data), type(crypto_data)
crypto_data = crypto_data[0] #Pega a primeira tabela do DataFrame

crypto_data
```

###Exibe o tipo de DataFrame

```python
import pandas as pd
import requests

type(crypto_data)
```

```python
import pandas as pd
import requests

crypto_final = crypto_data.iloc[:,2:] #Deleta as duas primeiras colunas
crypto_final.head()
```

###Trocando nome de algumas colunas

```python
crypto_final.columns = crypto_final.columns.str.replace('Last price','last_price')
crypto_final.columns = crypto_final.columns.str.replace('24 high', 'high')
crypto_final.columns = crypto_final.columns.str.replace('24 low','low')
crypto_final.columns = crypto_final.columns.str.replace('Price Charts 7d','price_charts')

crypto_final.head()
```
