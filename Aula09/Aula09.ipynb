# Trabalhando com Pandas: Average, Median and Mode
## Carregando DataSet (Parte-I)

```python
import pandas as pd

data = [['Cloudmagic', 12], ['Smartr', 17], ['Sunrise Calendar', 11],['Free-time', 16], ['Pocket', 16], ['Wunderlist', 12], ['Files Google', 18]]
df = pd.DataFrame(data, columns=['App','ms'])
df
```

## Carregando DataSet (Parte-II)

```python
import pandas as pd

data = {'App': ['Cloudmagic','Smartr', 'Sunrise Calendar','Free-time', 'Pocket','Wunderlist', 'Files Google'],'ms': [12, 17, 11, 16, 16, 12, 18]}
df = pd.DataFrame(data)
df
```

## Average or Mean


```python
df['ms'].mean()
```

## Median


```python

df['ms'].median()
```

##Mode


```python

df['ms'].mode()
```

## Estatística Descritiva

```python
df['ms'].describe()
```

##Seleções Condicionais

```python
#LOC aplica uma instrução em 1 coluna
#ILOC aplica uma instrução em várias colunas
result = df.loc[df['ms'] == 16]
result
```

```python
result = df.loc[df['ms'] >= 16]
result
```

```python
result = df.loc[df['ms'] >= 16]
result.plot.area(x='App') #Insere um gráfico
#barh, bar, line - tipos de gráficos
```
