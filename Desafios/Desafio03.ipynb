### Desafio 3 – Use o Dataframe: https://raw.githubusercontent.com/Simone-Ingrid/datasets/main/base_precos.csv .Considerando que o Python é Case sensitive, os dados únicos de uma coluna devem ser consertados, exemplo ‘brinquedo’ é a mesma coisa que ‘brinqueDo’. No desafio, deve-se consertar essas strings da coluna 'setor’.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Simone-Ingrid/datasets/main/base_precos.csv', sep=';', encoding='unicode_escape')
pd.set_option('display.max_rows', None)
df['setor'][0:146]='mesa_banho'
df
```
