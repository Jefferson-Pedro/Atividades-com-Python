```python
import pandas as pd
arquivo = 'analise_homicidios.csv'
ler = pd.read_csv(arquivo, sep=";")
display(ler)
```

```python
import csv

with open('analise_homicidios.csv', "r") as arquivo:
  ler = csv.reader(arquivo, delimiter=";")
  for i, linha in enumerate(ler):
    print(linha)
```

```python
import csv

with open('analise_homicidios.csv', "r", encoding="utf-8") as arquivo:
  analise = arquivo.readlines()
  for indice, linha in enumerate(analise):
    print(indice,'|', linha)
```

```python
palavras = str(analise).split(";")
for palavra in palavras:
  if '' in palavra:
    print(palavra[:-1])
print(palavra)
```
