## Estrutura de Dados com Pandas: Series

```python
import pandas as pd
notas = pd.Series([2,7,5,10,6,20])
print(notas)
print("\n",notas.index) #Atributos índices
print("\n",notas.values) #Atributos valores
print("\n",notas.head(n=3)) #Informações de 3 primeiras linhas de dados!
print("\n",notas.tail(5)) #Informações de 3 últimas linhas de dados!
print("\n",notas.shape) #Informações de quantidade de linhas e colunas!
print()
type(notas)
```

## Estruturas dos Dados: Dataframe

```python
a = [['a', '1.2', '4.2'], ['b', '70', '0.03'], ['x', '5', '0']]
df = pd.DataFrame(a, columns=['one', 'two', 'three'])
print(df)
```

## Converter os tipos de dados: DataFrame

```python
a = [['a', '1.2', '4.2'], ['b', '70', '0.03'], ['x', '5', '0']]
df = pd.DataFrame(a, columns=['one', 'two', 'three'])
df[['two', 'three']] = df[['two', 'three']].astype(float)
print(df.info())
```

# DataFrame: Trocar o nome das colunas do Dataframe.

```python
df.columns= ['id', 'data_aq', 'produto', 'quantidade', 'valor_un', 'valor_total', 'setor']
print(df.head(n=5))
```

## Métodos de Conversão

```python
print(df.to_string()) #Converte os dados das colunas em string
print(df.to_html()) #Converte os dados das colunas em HTML.
print(df.to_json()) #Converte os dados no formato json (JavaScript Object Notation).
print(df.to_dict()) #Converte os dados em dicionário (dict) Python.Necessita de outros parâmetros para ajustar as chaves.
```
