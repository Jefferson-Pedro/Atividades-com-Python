```python
import pandas as pd
arquivo = 'analise_homicidio.csv'
ler = pd.read_csv(arquivo, sep=";")
display(ler)
```

```python
import csv

with open('analise_homicidio.csv', "r") as arquivo:
  ler = csv.reader(arquivo, delimiter=";")
  for i, linha in enumerate(ler):
    print(linha)

```
