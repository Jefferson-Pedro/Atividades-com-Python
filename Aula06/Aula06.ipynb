## Criando arrays com NumPy

```python
import numpy as np
# list in Python 
data = [1, 2, 3, 4, 5, 6]
# list with parameter for Numpy array function 
array = np.array(data)
print(type (array)) 
print(array)
```

## Criando arrays com NumPy

```python
# nested list 
data = [[1, 2, 3], [4, 5, 6]]
# list 2D 
array = np.array(data)
print(type (array)) 
print(array)
```
