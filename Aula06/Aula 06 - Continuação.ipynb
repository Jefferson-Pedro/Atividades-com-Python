# **Gerando Arrays**

```python
import numpy as np

print(np.zeros(10))
print(np.ones(10))
print(np.arange(20+1))
```

## **Gerando arrays com Métodos Built-in**

```python
print(np.arange(0,10,1))
print(np.arange(0,10,2))
print(np.arange(0,10,3))
```

## **Arrays Aleatórios Unidimensional**



```python
print(np.random.rand(1))
print(np.random.rand(2))
print(np.random.rand(3))
```

## **Arrays Aleatórios Multidimensional**


```python
#Cria uma matriz
print(np.random.rand(2,2))
print(np.random.rand(2,4))
```

## Principais Atributos: Max() e Min()

```python
vetor = np.random.rand(3)
print(vetor)
print(vetor.max())
print(vetor.min())
```

## *Principais Atributos: dtype* 

```python
#Pra ver qual tipo de dado do vetor
vetor = np.random.rand(3)
print(vetor.dtype)
vetor = np.arange(3)
print(vetor.dtype)
```

## *Principais Atributos: Slicing*

```python
vetor = np.arange(1,20,1)
print(vetor)
vetor1 = vetor[1:6]
print(vetor1)
vetor[1:6] = 20
print(vetor)
```

## Principais Funções Matemáticas

```python
vetor = np.arange(4)
print(np.exp(vetor))
print(np.sqrt(vetor))
```

## Transformando Arquivo em Arrays: genfromtxt


```python
import numpy as np 
from numpy import genfromtxt

#importando arquivos para dentro do colab
data = genfromtxt("arquivos_aula06/array1.txt", delimiter = ",")
print(data)
```

## Transformando Arquivo em Arrays: loadtxt


```python
data = np.loadtxt("arquivos_aula06/array1.txt", dtype = int, delimiter=",")
print(data)
```
