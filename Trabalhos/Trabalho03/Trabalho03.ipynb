```python
import pandas as pd

url = 'https://www.w3resource.com/python-exercises/pandas/filter/world_alcohol.csv'
storage_options = {'User-Agent': 'Mozilla/5.0'}

df = pd.read_csv(url, storage_options=storage_options)
df.rename(columns={'Display Value': 'Display_Value'}, inplace = True)
pd.set_option('display.max_rows', None)
df
```

```python
#a. Agrupe os dados por tipo de bebidas;
g = df.groupby('Beverage Types').first()
g
```

```python
#b. Agrupe os dados por Região e por Ano;
pd.set_option('display.max_rows', None)
g = df.groupby(['WHO region', 'Year']).first()
g
```

```python
#c. Seção de Contagens: Contar a ocorrência de Regiões, de Países e a soma da coluna de valores por Bebida.
pd.set_option('display.max_rows', None)

r = df.groupby('WHO region')['WHO region'].count()
p = df.groupby('Country')['Country'].count()
s = df.groupby('Beverage Types')['Display_Value'].sum()
print('Ocorrencia por região:\n ', r )
print('-=' * 20)
print('Soma por tipo de bebida:\n ', s )
print('-=' * 20)
print('Ocorrencia por país:\n ', p )
```

```python
#d. Realize análises estatísticas da coluna dos valores: Média, Moda, Mediana, 
#Estatística Descritiva e Gráfico de comparação dos valores agrupados por tipo de bebida.

media = df['Display_Value'].mean()
moda = df['Display_Value'].mode()
mediana = df['Display_Value'].median()
ed = df.groupby('Beverage Types').describe()
print(f'Média:{media}, \nModa:{moda}, \nMediana:{mediana}')
ed
```

```python
#Gráfico de comparação dos valores agrupados por tipo de bebida.
query = df.groupby('Beverage Types')['Display_Value'].sum()
query.plot.bar(y = 'Comparação Valor x TipoBebida')
```

```python
#e. Mostre resultados de acordo com alguns critérios: i. Mostrar a coluna de bebidas do ano de 1985.
df.query('Year == 1985')['Beverage Types']
```

```python
#e. Mostre resultados de acordo com alguns critérios: ii. Mostrar a coluna de Região com valores acima de 4 

df.query('Display_Value > 4')['WHO region']
```
