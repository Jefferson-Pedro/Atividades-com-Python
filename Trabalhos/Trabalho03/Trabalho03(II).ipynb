```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Jefferson-Pedro/BigData-com-Python/main/Trabalhos/Trabalho03/cursos-prouni.csv'
df = pd.read_csv(url, encoding='utf-8')
df
```

```python
#a. Efetuar a limpeza das colunas de notas: onde tiver NaN (Not a Number), substituir por 0,0.
df.fillna(0, inplace=True)
df
```

```python
#b. Agrupe os dados pelo grau (Bacharelado, Licenciatura, etc).
gb = df.groupby('grau').first()
gb
```

```python
#c. Agrupe os dados pelos cursos de Matemática, Medicina e Pedagogia.
g = df.loc[(df['curso_busca'] == 'Medicina') 
| (df['curso_busca'] == 'Pedagogia') 
| (df['curso_busca'] == 'Matemática')].groupby('curso_busca').max()
g
```

```python
#d. Agrupe os dados por Estado.
gb = df.groupby('uf_busca').first()
gb
```

```python
#e. Agrupe os dados pelos cursos Tecnológicos.    .agg(list)
gb = df.loc[(df['grau'] == 'Tecnológico')].groupby('curso_busca').first()
gb
```

```python
#f. Elimine a coluna “cidade_filtro” do dataframe.
gb = df.drop(columns=['cidade_filtro'])
gb
```

```python
#g. Apresente a média das mensalidades dos cursos de Medicina.
gb = df.loc[(df['curso_busca'] == 'Medicina')].groupby('turno')['mensalidade'].mean()
pd.DataFrame(gb)
```

```python
#h. Média das notas de corte dos cursos de tempo integral.
gb = df.loc[(df['turno'] == 'Integral')]
listaNotas = gb.groupby('curso_busca')[['nota_integral_ampla','nota_integral_cotas','nota_parcial_ampla','nota_parcial_cotas']].mean()
pd.DataFrame(listaNotas)
```

```python
#i. Estatística Descritiva das Notas Integral Ampla dos cursos de Bacharelado.
gb = df.loc[df['grau'] == 'Bacharelado'].groupby('curso_busca').describe()
gb
```

```python
#j. Gráfico comparativo entre o grau dos cursos (Bacharelado, Licenciatura, Tecnologia, etc) pelas Notas Integral de Cotas.
query = df.groupby('grau')['nota_integral_cotas'].sum()
query.plot(kind='bar',title='Grau x Notas' , ylabel = 'Notas Integral de Cotas', figsize =(15,10))
```
